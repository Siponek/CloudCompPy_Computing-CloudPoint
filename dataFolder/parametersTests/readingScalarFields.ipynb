```python
from datetime import datetime
import csv

import laspy

print(datetime.now(), "Loaded libraries")
```

```python
path_to_scalar_value_file : str = "".join(["./ParametersConclusion/", "conclusion_M3C2_scalar_values", ".csv"])
path_to_uncertainty_value_file : str = "".join(["./ParametersConclusion/", "conclusion_M3C2_uncertainty_values", ".csv"])

rows_for_csv_M3C2_distances : list = []
rows_for_csv_uncertainty : list = []


with open(path_to_uncertainty_value_file, "w+") as file_uncertainty:
    with open(path_to_scalar_value_file, "w+") as file_scalar_value:
        csvwriter_scalar = csv.writer(file_scalar_value, lineterminator="\n")
        csvwriter_uncertainty = csv.writer(file_uncertainty, lineterminator="\n")
        scalar_fields_for_csv = ['M3C2_scalar_values', 'Min', 'Max']
        uncertainty_fields_for_csv = ['M3C2_distance_uncertainty_values', 'Min', 'Max']
        csvwriter_scalar.writerow(scalar_fields_for_csv)
        csvwriter_uncertainty.writerow(uncertainty_fields_for_csv)

        for majorIndex in range(1, 4):
            for index in range(1, 31):
                print(index, "of", 30)
            # output_1_1.las
                path_to_las_input:str = "".join(["output_", str(majorIndex), "_", str(index), ".las"])
                # path_to_las_input:str = f"output_{majorIndex}_" + str(index) + ".las"
                # print(datetime.now(), "Path to las input:", path_to_las_input)
                # print(datetime.now(), "Path to conclusion file:", path_to_scalar_value_file)
                with laspy.open(path_to_las_input) as file:
                    las = laspy.read(path_to_las_input)
                    try:
                        print(datetime.now(), f"Reading M3C2_distance from {path_to_scalar_value_file}")
                        m3c2_distance_field = las.M3C2_distance
                        m3c2_distance_field_min = min(m3c2_distance_field)
                        m3c2_distance_field_max = max(m3c2_distance_field)
                        rows_for_csv_M3C2_distances = [path_to_las_input, m3c2_distance_field_min, m3c2_distance_field_max]

                        #* this part can be failed if try statment throws an error before, but distance uncertainty cannot exist without M3C2_distance in the first place
                        print(datetime.now(), f"Reading uncertainty_value from {path_to_uncertainty_value_file}")
                        uncertainty_value = las.distance_uncertainty
                        uncertainty_value_min = min(uncertainty_value)
                        uncertainty_value_max = max(uncertainty_value)
                        rows_for_csv_uncertainty = [path_to_las_input, uncertainty_value_min, uncertainty_value_max]
                        
                        
                        # file_scalar_value.write(f"M3C2_scalar_values of {path_to_las_input} \t\tfrom\t{m3c2_distance_field_min}\t\t\t to\t{m3c2_distance_field_max}\n")
                    except Exception as e:
                        print(datetime.now(), "Error reading M3C2_distance")
                        print(datetime.now(), e)
                        rows_for_csv_M3C2_distances = [[path_to_las_input, str(e)]]
                        rows_for_csv_uncertainty = [[path_to_las_input, str(e)]]
                        # file_scalar_value.write(f"M3C2_scalar_values of {path_to_las_input} \t\terror : {str(e)}\n")
                    finally:
                        # print("Closing file:", path_to_las_input)
                        # print("Writting to file:", path_to_scalar_value_file)
                        csvwriter_scalar.writerow(rows_for_csv_M3C2_distances)
                        csvwriter_uncertainty.writerow(rows_for_csv_uncertainty)

```
